# Handling Date and Time

```python
date = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

```python
date.head()
```

```python
time.head()
```

```python
date.info()
```

```python
time.info()
```

```python
# Treatment of date and time
#  - date
# Convert date object into date time dtype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extract date year
date['year'] = date['date'].dt.year
```

```python
date.sample(5)
```

```python
# Extract Month
date['month'] = date['date'].dt.month
date.head()
```

```python
# Month Name
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# Extract day
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day number of the week
date['dayNum_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# day name
date['day_Name_week'] = date['date'].dt.day_name()
date.head()
```

```python
# Is weekend?
date['date_is_weekend'] = np.where(date['day_Name_week'].isin(['Sunday', 'Saturday']), 1,0)
date.head()
```

```python
# week number
date['date_week'] = date['date'].dt.week
date.head()
```

```python
# quater number
date['Quarter'] = date['date'].dt.quarter
date.head()
```

```python
# Semester number
date['Semester'] = np.where(date['Quarter'].isin([1,2]),1,2)
date.head()
```
